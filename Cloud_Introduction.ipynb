# CLOUD COMPUTING

## 1.1- What is a Cloud?

A cloud refers to a network of remote servers hosted on the internet to store, manage, and process data, rather than on a local server or personal computer. Cloud services provide computing power, storage, and applications on demand.

## 1.2. Why Clouds?

1. __Cost Efficiency:__ Reduced capital expenditure as users only pay for the resources they consume.
2. __Scalability and Flexibility:__ Easily scale resources up or down based on demand.
3. __Accessibility:__ Access resources and applications from anywhere with an internet connection.
4. __Maintenance:__ Reduced need for in-house IT maintenance and management.
5. __Innovation:__ Faster deployment of new services and features.

## 1.4. Advantages of Cloud Computing

1. __Cost Efficiency:__ Reduced capital expenditure as you pay only for the resources you use, avoiding the costs of hardware purchase and maintenance.

2. __Scalability and Flexibility:__ Easily scale resources up or down based on demand without significant delays.

3. __Accessibility:__ Access resources and applications from anywhere with an internet connection.

4. __Maintenance and Management:__ Reduced need for in-house IT maintenance as the cloud provider manages the infrastructure.

5. __Innovation and Speed:__ Faster deployment of new services and features, accelerating innovation.

6. __Security and Compliance:__ Cloud providers offer advanced security features and compliance certifications.

## 1.5. Disadvantages of Cloud Computing

1. __Downtime:__ Service outages or downtime can impact business operations, especially if the cloud provider experiences technical issues.

2. __Limited Customization:__ Standardized offerings may not meet all specific business needs, limiting customization options

3. __Vendor lock-in:__ Situation where a customer becomes dependent on a single cloud service provider for products and services, making it difficult to switch to another provider without substantial costs, effort, or disruption.

4. __Security Risks:__ When companies use cloud providers for storing and processing their data, they give a third-party provider access to their sensitive information. 

## 1.6.  Cloud IS A Distributed System

A cloud is inherently a distributed system, meaning it consists of multiple interconnected computers that work together to appear as a single system to the user.

A distributed system is a network of independent computers that communicate and coordinate with each other to achieve a common goal. These systems are characterized by:

Concurrency: Multiple processes run simultaneously.
Scalability: The system can grow and manage increased demand.
Fault Tolerance: The system can continue to function even when parts of it fail.
